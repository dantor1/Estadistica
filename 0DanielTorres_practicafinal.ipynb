Miremos el código postal. Es una variable con entradas erroneas.
Hay valores como '', '-'  y '28' que deberían ser considerados como NA.
Así mismo también debería ser NA todos los que no comiencen por 28, ya que estamos con códigos postales de Madrid.

El código postal 28002, 28004 y 28051 tienen entradas repetidas. 
Por ejemplo las entradas 28002\n20882 deberían ir dentro de 28002

El codigo 2804 debería ser 28004, 2805 deberia ser 28005 y 2815 junto con 2815 debería ser 28015

Limpia los datos de la columna Codigo Postal

```r
airbnb$CodigoPostal[airbnb$CodigoPostal == ''] = NA
airbnb$CodigoPostal[airbnb$CodigoPostal == '-'] = NA
airbnb$CodigoPostal[airbnb$CodigoPostal == '28'] = NA

airbnb$CodigoPostal[airbnb$CodigoPostal == '28002\n28002'] = '28002'
airbnb$CodigoPostal[airbnb$CodigoPostal == 'Madrid 28004'] = '28004'
airbnb$CodigoPostal[airbnb$CodigoPostal == '28051\n28051'] = '28051'

airbnb$CodigoPostal[airbnb$CodigoPostal == '2804'] = '28004'
airbnb$CodigoPostal[airbnb$CodigoPostal == '2805'] = '28005'
airbnb$CodigoPostal[airbnb$CodigoPostal == '2815'] = '28015'

airbnb$CodigoPostal[airbnb$CodigoPostal == '280013'] = '28013'
airbnb$CodigoPostal[airbnb$CodigoPostal == '2015'] = '28015'


airbnb$CodigoPostal[!grepl("^28",airbnb$CodigoPostal)] = NA
#airbnb[which(!grepl("^28",airbnb$CodigoPostal)),] = NA


airbnb$CodigoPostal = factor(airbnb$CodigoPostal)
table(airbnb$CodigoPostal)
```

Ayuda: Usa la función aggregate `aggregate(.~colname,df,mean,na.rm=TRUE)`

¿Cuantos apartamentos hay en cada uno de esos barrios?

Mostrar una dataframe con el nombre del barrio, el precio y el número de entradas.

Ayuda: Podeis crear un nuevo dataframe con las columnas "Barrio" y "Freq" que contenga el número de entradas en cada barrio y hacer un merge con el dataframe del punto anterior.

```r
df_entire = data.frame(table(airbnb_entire$Barrio))
colnames(df_entire) = c("Barrio","Freq")
newdf = merge(df_entire, df, by="Barrio")
top5precio = newdf[order(newdf$Precio, decreasing = TRUE),]
head(top5precio, 5)
#top5precio = top5precio[order(top5precio$Precio, decreasing = T),]
```

Partiendo del dataframe anterior, muestra los 5 barrios con mayor precio, pero que tengan más de 100 entradas de alquiler.

```r
Freqmayor100 = top5precio[top5precio$Freq > 100,]
head(Freqmayor100,5)
top5barrio100 = head(Freqmayor100,5)$Barrio
```

Dibuja el diagrama de densidad de distribución de los diferentes precios. Serían 5 gráficas, una por cada barrio.

```r
library(ggplot2)
options(repr.plot.height=4, repr.plot.width=6, repr.plot.res = 300)


densidadesbarrios = subset(airbnb_entire,Barrio %in% top5barrio100)
ggplot(densidadesbarrios, aes(x=Precio, color=Barrio))+ 
  geom_density(aes(x = Precio, fill = Barrio))+ 
  facet_grid(Barrio~.)+
  ggtitle("densidad de distribución de los diferentes precios")
```

Calcula el tamaño medio, en metros cuadrados, para los 5 barrios anteriores y muestralo en el mismo dataframe junto con el precio y número de entradas

```r
tamanyomedio = aggregate(. ~ Barrio, airbnb_entire[c("Barrio","MetrosCuadrados")], mean, na.rm=TRUE)
tamanyomedio2 = merge(Freqmayor100, tamanyomedio, by="Barrio")
tamanyomedio2 = tamanyomedio2[order(tamanyomedio2$Precio, decreasing = TRUE),]
head(tamanyomedio2, 5)
```

Dibuja el diagrama de densidad de distribución de los diferentes tamaños de apartamentos. Serían 5 gráficas, una por cada barrio.

```r
ggplot(data=densidadesbarrios, aes(x=MetrosCuadrados, color=Barrio))+
geom_density(aes(x=MetrosCuadrados, fill=Barrio), position = 'stack')+
facet_grid(Barrio~., scales = 'free')
```

Esta claro que las medias de metros cuadrados de cada uno de estos 5 barrios parecen ser diferentes, pero ¿son estadísticamente diferentes?
¿Que test habría que usar para comprobarlo?

```r
for (barrio in unique(densidadesbarrios$Barrio)){
    df = densidadesbarrios[which(densidadesbarrios$Barrio==as.character(barrio)),]
    print(paste(barrio, round(mean(df$MetrosCuadrados, na.rm=TRUE), 2)))
}
```

¿Cual es la probabilidad de encontrar, en el barrio de Sol, un apartamento en alquiler con 3 dormitorios?
¿Cual es el intervalo de confianza de esa probabilidad?

```r
print("probabilidad de encontrar, en el barrio de Sol, un apartamento en alquiler con 3 dormitorios:")
sum(barrio_sol$NumDormitorios == 3)/nrow(barrio_sol)
print("el intervalo de confianza de esa probabilidad")
binom.test(sum(barrio_sol$NumDormitorios == 3), nrow(barrio_sol))
```
